```python
from bitpack import BPDR
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
iris_data = iris.data
iris_targets = iris.target
```

```python
print(iris_data[0])
```

```python
bpdr = BPDR(n_components=2)
```

```python
bpdr_data = bpdr.fit_transform(iris_data, iris_targets)
```

```python
bpdr_data[0]
```

```python
print(bpdr.variances)
print(bpdr.scaled_variances)
print(bpdr.mean_variance)
```

```python
x1 = []
x2 = []

for row in bpdr_data:
    x1.append(row[0])
    x2.append(row[1])
    
plt.scatter(x1, x2, c=iris_targets, cmap=plt.cm.Set1, edgecolor='k')
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(iris_data)
```

```python
x3 = []
x4 = []

for row in principalComponents:
    x3.append(row[0])
    x4.append(row[1])

plt.scatter(x3, x4, c=iris_targets, cmap=plt.cm.Set1, edgecolor='k')
plt.show()    
```
